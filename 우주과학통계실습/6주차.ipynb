3개국의 남성 평균신장의 정규분포가 아래와 같을 때 이를 그려보라
1. 평균 170cm, 표준편차 10cm
2. 평균 170cm, 표준편차 5cm
3. 평균 175cm, 표준편차 10cm

```python
from scipy.stats import norm
import scipy
```

```python
data = {'country1':[170,10], 'country2':[170,5], 'country3':[175,10]}
for i, (key, items) in enumerate(data.items()):
    
    avg = items[0]
    std = items[1]
    x_values = np.arange(avg - 5*std, avg+5*std, 0.1)
    y_values = norm(loc = avg, scale=std).pdf(x_values)
    plt.plot(x_values, y_values, label = key)
```

```python
t = np.arange(0,4.25, 0.25)
x2 = t/np.sqrt(2)
print(scipy.special.erf(x2))
```

실습 : 정규분포를 따르는 시험 결과

```python
x_mean = 50
x_sd = 15

x_rv = norm(loc=x_mean, scale = x_sd)
x_values = np.linspace(x_mean - 4*x_sd, x_mean+4*x_sd, 100)
y_values = x_rv.pdf(x_values)
plt.plot(x_values, y_values, lw=2)
```

```python
x_cut = 40
p_cut = x_rv.cdf(np.inf) - x_rv.cdf(x_cut)
print(p_cut)
```

```python
x_rv.ppf(0.8)
```

```python
boot_avg = 7
boot_std = 0.5
boot = norm(loc = boot_avg, scale = boot_std)
boot.ppf(0.01)
```
